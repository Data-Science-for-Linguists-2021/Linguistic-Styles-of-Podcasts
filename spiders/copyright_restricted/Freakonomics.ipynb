```python
import scrapy
from scrapy.crawler import CrawlerProcess
```

```python
freakonomics_data = []
```

```python
# parses transcript links
class Freakonomics(scrapy.Spider):
    name = 'Freakonomics'
    
    def start_requests(self):
        url = 'https://freakonomics.com/archive/'
        yield scrapy.Request(url=url, callback=self.parse_episode)
        
    def parse_episode(self, response):
        transcript_links = response.xpath('//tr//span[@class="green-title"]/a/@href').extract() 
        # print([l for l in transcript_links])
        for link in transcript_links:
            yield scrapy.Request(url=link, 
                                  callback =self.parse_transcript)
    
    def parse_transcript(self, response):
        current_ep = []
        
        title = str(response.xpath('//h1[@class="single_title"]/text()').extract())
        clean_title = title[2:-3].split(' (Ep. ')
        ep_number = clean_title[1]
        current_ep.append(ep_number)
        
        date = str(response.xpath('//div[@class="article_meta"]//time/text()').extract())
        current_ep.append(date)
        
        ep_title = clean_title[0]
        current_ep.append(ep_title)
        
        text = (response.css(' div.podcast_intro p > *::text').extract())[2:]
        current_ep.append(text)
        
        freakonomics_data.append(current_ep)
        
        
```

```python
process = CrawlerProcess()
process.crawl(Freakonomics)
process.start()
```

```python
import pandas as pd
freak_df = pd.DataFrame(freakonomics_data,
                        columns = ['Episode','Year','Title','Text'])
freak_df.head()
# Looks like i only need to clean the year and the text
```

```python
import re

regex = r'[0-9]{4}(?= @)' # any non whitespace character between a comma and an @
freak_df['Year'] = freak_df.Year.map(lambda d: re.findall(regex, d)[0])
freak_df['Year'][:10]
```

```python
# start from index 3, get rid of \xa0, and join
freak_df.Text[0]
```

```python
freak_df['Text'] = freak_df.Text.map(lambda t: str('  '.join(t[3:])).replace('\xa0', ''))
freak_df.Text[0]
```

```python
def ep_sort(episode):
    while len(episode) < 3:
        episode = '0' + episode
    return episode

ep_sort('04')
```

```python
# get rid of rebroadcasts
freak_df = freak_df[~freak_df.Episode.str.contains('Rebroadcast')]
freak_df
    
```

```python
freak_df['Episode'] = freak_df.Episode.map(lambda e: ep_sort(e))
freak_df = freak_df.sort_values(by=['Episode'])
freak_df.Episode.values
```
