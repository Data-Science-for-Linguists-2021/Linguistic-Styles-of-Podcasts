```python
import scrapy
from scrapy.crawler import CrawlerProcess
import re
```

```python
OnBeing_data = []
```

```python
# parses transcript links
class On_Being(scrapy.Spider):
    name = 'OnBeing'
    
    def start_requests(self):
        base_url = 'https://onbeing.org/series/page/'
        urls = []
        for i in range(1, 20):
            urls.append(base_url + str(i) + '/')
        # print(urls) # working
        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse_episode)
        
    def parse_episode(self, response):
        transcript_links = response.xpath('//a[contains(@class,"transcript")]/@href').extract() 
        # print([l for l in transcript_links]) # working
        for url in transcript_links:
            yield response.follow(url=url, 
                                  callback =self.parse_transcript)
    
    def parse_transcript(self, response):
        current_ep = []
        # Title - done, clean
        title = response.xpath('//div[@class="episode__header-titles"]/*/text()').extract()
        title = ': '.join(title)
        current_ep.append(title)
        
        # No episode number
        date = response.css(' p.episode__published-date::text').extract()
        current_ep.append(date[-1])
        
        text = response.css(' div.episode__transcript-body-inner > p').extract()
        current_ep.append(text)
        
        OnBeing_data.append(current_ep)
        
```

```python
process = CrawlerProcess()
process.crawl(On_Being)
process.start()
```

```python
import pandas as pd

OnBeing_df = pd.DataFrame(OnBeing_data,
                          columns = ['Title','Year','Text'])
OnBeing_df
```

Data looks pretty good.  Title doesn't need cleaned, but the year needs extracted from the date column.

```python
import re

OnBeing_df['Year'] = OnBeing_df.Year.map(lambda d: re.findall(r'\d{4}', d)[0])
```

```python
# Get rid of ugly tags and whatever xa0 is.  Keeping the \n's because they're sugnificant for the poems
import re

tags = r'<\S+>'
OnBeing_df['Text'] = OnBeing_df.Text.map(lambda t: '  '.join([word for word in t if len(word) > 0]).replace('\xa0', ''))
OnBeing_df['Text'] = OnBeing_df.Text.map(lambda t: re.sub(tags, '', t))
OnBeing_df.Text[0]
```
