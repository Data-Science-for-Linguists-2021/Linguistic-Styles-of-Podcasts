```python
import scrapy
from scrapy.crawler import CrawlerProcess
import re
```

```python
Invisible_data = []
```

```python
# parses transcript links
class Invisible(scrapy.Spider):
    name = '99%-Invisible'
    
    def start_requests(self):
        base_url = 'https://99percentinvisible.org/episodes/page/'
        urls = [(base_url + str(i)) for i in range(2, 30)] # Not doing as many pages because the website is very slow
        # print(urls) # Urls are correct
        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse_episode)
        
    def parse_episode(self, response):
        transcript_links = response.xpath('//a[@class="transcript"]/@href').extract() 
        # print([l for l in transcript_links])
        for link in transcript_links:
            link = 'https://99percentinvisible.org' + link
            yield response.follow(url=link, 
                                  callback =self.parse_transcript)
    
    def parse_transcript(self, response):        
        # Title - done, clean
        current_ep = []
        title = response.xpath('//h1[@class="entry-title"]/text()').extract()[0]
        current_ep.append(title)
        
        date = response.xpath('//div[@class="meta-date"]/p[@class="meta-value"]/time/text()').extract()[0]        
        current_ep.append(date)
        
        number = response.xpath('//h4[contains(@class,"entry-label")]/text()').extract()[0]
        current_ep.append(number)
        
        
        text = response.xpath('//div[contains(@class, "transcript-content")]/*//text()').extract()
        current_ep.append(text)
        
        Invisible_data.append(current_ep)
```

```python
process = CrawlerProcess()
process.crawl(Invisible)
process.start()
```

```python
# Make dataframe from List of lists (Invisible_data)
import pandas as pd

Invisible_df = pd.DataFrame(Invisible_data,
                            columns = ['Title','Date','Episode','Text'])
```

```python
Invisible_df
```

```python
# Titles look good, but everything else needs cleaned.  Starting with date.
Invisible_df['Year'] = Invisible_df.Date.map(lambda date: '20' + date[-2:])
Invisible_df.head()
```

```python
# Now to clean the episode numbers - can't use regex because some of the numbers didn't come through
Invisible_df['Episode'] = Invisible_df.Episode.map(lambda e: ''.join([n for n in e if n.isnumeric()]))
Invisible_df.head()
```

```python
# Now for the hard part, the text . . . 
Invisible_df.Text[0]
```

```python
# Don't quite need any regex just yet
Invisible_df['Text'] = Invisible_df.Text.map(lambda t: ' '.join(t).replace('\n','').replace('\xa0',''))
Invisible_df.Text[0]
```

```python
# Fix columns
Invisible_df
```

```python
Episode = Invisible_df.pop('Episode')
Title = Invisible_df.pop('Title')
Year = Invisible_df.pop('Year')
Text = Invisible_df.pop('Text')
Invisible_df = Invisible_df.drop(['Date'], axis=1)

Invisible_df
```

```python
Invisible_df['Episode'] = Episode
Invisible_df['Title'] = Title
Invisible_df['Year'] = Year
Invisible_df['Text'] = Text

Invisible_df.head()
```
