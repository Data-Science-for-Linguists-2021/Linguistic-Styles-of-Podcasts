```python
def parse_ira(text):
    
    text = ' -- '.join(text)    
    ira_speech = re.findall(r'(?<= -- Ira Glass -- ).*?(?= -- [A-Z][a-z]+)', text)
    ira_speech = [s.strip() for s in ira_speech]
    ira_speech = '  '.join(ira_speech)
    return ira_speech

parse_ira(sample)
```

```python
tal_df['Ira_speech'] = tal_df.Text.map(parse_ira)
tal_df
```

```python
import spacy
from spacy.lang.en import English


nlp = spacy.load('en_core_web_md')
```

```python
# tokenize Ira Glass's speech only
tal_df['Tokens'] = tal_df.Ira_speech.map(lambda s: nlp(s.lower()))
tal_df.sample(5)
```

```python
# add token count (transcript_length) column
tal_df['Speech_len'] = tal_df.Ira_speech.map(len)
tal_df
```

```python
# delete rows that didn't read in speech
tal_df = tal_df[tal_df.Speech_len != 0]
tal_df.info()
```

```python
from collections import Counter
all_stopwords = nlp.Defaults.stop_words

def top10(Tokens):
    counts = Counter(t.text for t in Tokens if t.is_alpha and t not in all_stopwords)
    return counts.most_common(10)

top10(tal_df.Tokens[3])
```

```python
tal_df['Top_10'] = tal_df.Tokens.map(top10)
tal_df.Top_10.sample(3)
```
