```python
import scrapy
from scrapy.crawler import CrawlerProcess
import re
```

```python
eps = []
titles = []
year = []
content = []
```

```python
# parses transcript links
class On_Being(scrapy.Spider):
    name = 'OnBeing'
    
    def start_requests(self):
        base_url = 'https://onbeing.org/series/page/'
        urls = []
        for i in range(1, 20):
            urls.append(base_url + str(i) + '/')
        # print(urls) # working
        for url in urls:
            yield scrapy.Request(url=url, callback=self.parse_episode)
        
    def parse_episode(self, response):
        transcript_links = response.xpath('//a[contains(@class,"transcript")]/@href').extract() 
        # print([l for l in transcript_links]) # working
        for url in transcript_links:
            yield response.follow(url=url, 
                                  callback =self.parse_transcript)
    
    def parse_transcript(self, response):        
        # Title - done, clean
        title = response.xpath('//div[@class="episode__header-titles"]/*/text()').extract()
        title = ': '.join(title)
        titles.append(title)
        
        # No episode number
        # No date
        
        text = response.css(' div.episode__transcript-body-inner > p::text').extract()
        content.append(text)
        
```

```python
process = CrawlerProcess()
process.crawl(On_Being)
process.start()
```

```python
import pandas as pd

OnBeing_df = pd.DataFrame([titles, content]).T
OnBeing_df.columns = ['Title','Text']
OnBeing_df
```

```python
OnBeing_df.Text.map(lambda t: ' '.join(t))
OnBeing_df.Text.map(lambda c: [t for t in c if t != '[' or t != ']' or t != '[]'])
OnBeing_df.Text.map(lambda c: ' '.join(c))
```
