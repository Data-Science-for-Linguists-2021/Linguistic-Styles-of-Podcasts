```python
maxfun_pdfdf = pd.DataFrame(pdf_data,
                            columns = ['Title','Text'])
maxfun_pdfdf    
```

```python
# need to extract podcast name from title
maxfun_pdfdf.Title.iloc[:10]
```

```python
# create function that splits title into a list of podcast, episode, title

def splitter(t):
    ep_title = []
    if 'TAZ' in t:
        t = t.replace('TAZ', 'The Adventure Zone Ep-')
    if 'TGG' in t:
        t = t.replace('TGG', 'The Greatest Generation')
    if 'FF' in t:
        t = t.replace('FF', 'Friendly-Fire')
    if 'FriendlyFire' in t:
        t = t.replace('FriendlyFire', 'Friendly-Fire')
    t = re.split(r'[-Ee]p\.*-?(?=\d)', t)
    ep_title.append(t[0])
    
    t = re.split(r'(?<=\d{3})[-_]?(?=\S)', t[1])
    ep_title.append(t[0])
    ep_title.append(t[1])
    ep_title = [t.lower() for t in ep_title]
    
    return ep_title

splitter('TAZ167-Graduation24-With-Frenemies-Like-This')
```

```python
maxfun_pdfdf['Title'] = maxfun_pdfdf.Title.map(splitter)
maxfun_pdfdf.head()
```

```python
# extract podcast title from splitter function run on title string
podcasts = [re.sub(r'[-_]', ' ', t[0]).strip() for t in maxfun_pdfdf.Title]
podcasts[:10]
```

```python
# extract episode number from splitter function run on title string
episodes = [re.sub(r'[-_]', ' ', t[1]).strip() for t in maxfun_pdfdf.Title]
episodes[:10]
```

```python
# extract episode title from splitter function run on title string
titles = [re.sub(r'[-_]', ' ', t[2]).strip() for t in maxfun_pdfdf.Title]
titles[:10]
```

```python
# rearrange columns 
maxfun_pdfdf = maxfun_pdfdf.drop(['Title'], axis=1)
contents = maxfun_pdfdf.pop('Text')
maxfun_pdfdf
```

```python
# re-build dataframe with columns in the appropriate position
maxfun_pdfdf['Podcast'] = podcasts
maxfun_pdfdf['Episode'] = episodes
maxfun_pdfdf['Title'] = titles
maxfun_pdfdf['Year'] = ''
maxfun_pdfdf['Text'] = contents

maxfun_pdfdf
```

```python
def getdate(text):
    date = re.findall(r'\d{4}?(?=[ ]\n|\\n)', text)
    if len(date) < 1:
        date.append('')
    year = date[0]
    
    return year
        
getdate(maxfun_pdfdf.Text[60])
maxfun_pdfdf.Text[60]
```

```python
maxfun_pdfdf['Year'] = maxfun_pdfdf.Text.map(getdate)
maxfun_pdfdf.head()
```
