```python
import scrapy
from scrapy.crawler import CrawlerProcess
from bs4 import BeautifulSoup
import requests
import re
```

```python
allusionist_data = []
```

```python
class Allusionist(scrapy.Spider):
    name = 'allusionist'
    
    def start_requests(self):        
        url = 'https://www.theallusionist.org/transcripts/category/transcript'
        # this one is tricky because the offset changes for every page
        pages = []
        keep_looping = True
        while keep_looping:
            r = requests.get(url)
            soup = BeautifulSoup(r.text, 'html.parser')
            url_tag = soup.find('a', class_='older-posts')
            if not url_tag:
                keep_looping = False
            else:
                url = 'https://www.theallusionist.org' + url_tag.get('href')
                pages.append(url)
        for link in pages:
            yield scrapy.Request(url=link, callback=self.parse_episode)
                                
    def parse_episode(self, response):
        transcript_links = response.xpath('//a[@class="read-more"]/@href').extract()
        for link in transcript_links:
            link = 'https://www.theallusionist.org' + link
            yield response.follow(url=link, 
                               callback =self.parse_transcript)

    def parse_transcript(self, response):
        current_ep = []
        
        title = response.xpath('//a[contains(@href, "/transcripts/")]/text()').extract()[0]
        
        episode_re = r'(?<=Allusionist )(\d+){1}'
        episode = re.findall(episode_re, title)[0]
        current_ep.append(episode)
        
        title = re.sub(episode_re, 'SPLIT', title).split('SPLIT')[-1]
        title = re.findall(r'(?<=[:.] )(.*?)(?= - transcript)', title)[0]
        current_ep.append(title)
               
        year = response.xpath('//time[@itemprop="datePublished"]/text()').extract()[0]
        year = re.findall(r'\d+$', year)[0]
        current_ep.append(year)
        
        text = response.xpath('//div[@data-layout-label="Post Body"]//text()').extract()
        text = '\n'.join(text)
        current_ep.append(text)
        
        allusionist_data.append(current_ep)
```

```python
process = CrawlerProcess()
process.crawl(Allusionist)
process.start()
```

```python
allusionist_data[:5]
# each episode is stored as a list: [episode #, episode title, year released, transcript text]
```

```python
# convert episode data to dataframe
import pandas as pd

allusionist_df = pd.DataFrame(allusionist_data, 
                   columns=['Episode','Title','Year','Text'])
allusionist_df
```

```python
allusionist_df['Title'] = allusionist_df.Title.map(lambda t: t.lower())
allusionist_df.sample(10)
```

```python
# create function that accounts for mismatched number length to allow for sorting by episode

def clean_ep(episode):
    while len(episode) < 3:
        episode = '0' + episode
    return episode

clean_ep('17')
```

```python
allusionist_df['Episode'] = allusionist_df.Episode.map(clean_ep)
allusionist_df.head()
```

```python
# get rid of podcast-specific words that have no linguistic value
replaced_strings = ['helen','zaltzman','allusionist','theallusionist.org','radiotopia']
```

```python
# remove \n's
def cleantext(text):
    text = text.split('\n')
    text = ' '.join(text)
    text = text.lower()
    for t in replaced_strings:
        text = text.replace(t, 'REDACTED')
    return text
    
cleantext(allusionist_df.Text[0][:1000])
```

```python
allusionist_df['Text'] = allusionist_df.Text.map(cleantext)
allusionist_df.head()
```

```python
allusionist_df = allusionist_df.sort_values(by=['Episode'])
allusionist_df.head()
```
